```python
import pandas as pd
import matplotlib.pyplot as plt
import platform
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic', size=13)
else:
    plt.rc('font', family='AppleGothic', size=8)
```

```python
df = pd.read_csv('meel.csv')
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
df=df.sort_values('DATE')
```

```python
df=df.set_index('DATE')
df
```

```python
plt.figure(figsize=(19,6))
plt.rc('figure',titlesize=25)
plt.title(f'2019~2022               : ')
plt.plot(df, color='blue', marker='o',label='')
plt.legend()
plt.savefig('meelchart.png')
plt.show()
```
