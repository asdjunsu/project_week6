```python
import pandas as pd
import matplotlib.pyplot as plt
import platform
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic', size=13)
else:
    plt.rc('font', family='AppleGothic', size=8)
```

```python
df = pd.read_csv('gas_UK.csv')
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
df=df.sort_values('DATE')
df=df.set_index('DATE')
df
```

```python
plt.figure(figsize=(19,6))
plt.rc('figure',titlesize=25)
plt.title(f'2019~2022년 천연가스 선물시장 가격 변화   국가 : 영국')
plt.plot(df, color='red', marker='o',label='가격')
plt.legend()
plt.savefig('gaschart.png')
plt.show()
```
