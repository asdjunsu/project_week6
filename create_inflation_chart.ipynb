```python
import pandas as pd
import matplotlib.pyplot as plt
import platform
import csv
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic', size=13)
else:
    plt.rc('font', family='AppleGothic', size=8)
```

```python
f=open('inflation.csv', encoding='utf-8-sig')
data=csv.reader(f)
```

```python
usa=[]
kor=[]
date=[]

for row in data:
    if 'USA' in row[0]:
        usa.append(round(float(row[2]),2))
        date.append(row[1])
    elif 'KOR' in row[0]:
        kor.append(round(float(row[2]),2))
f.close()
```

```python
plt.figure(figsize=(20,8))
plt.rc('figure',titlesize=25)
plt.title(f'2019~2022년 물가상승률 변화         단위 : 분기')
plt.plot(date,usa,color='blue',marker='o', label='미국')
plt.plot(date,kor,color='red',marker='o',label='한국')
for i, v in enumerate(date):
    plt.text(v, kor[i], kor[i], fontsize = 15,color='black',horizontalalignment='center',
             verticalalignment='bottom') 
for i, v in enumerate(date):
    plt.text(v, usa[i], usa[i],fontsize = 15,color='black',horizontalalignment='center', 
             verticalalignment='bottom') 
plt.legend(loc=2)
plt.savefig('inflation_chart.png')
```
