```python
import pandas as pd
import matplotlib.pyplot as plt
import platform
import csv
import pandas as pd

if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic', size=13)
else:
    plt.rc('font', family='AppleGothic', size=8)
```

```python
df1 = pd.read_csv('FEDFUNDS.csv')
df1['DATE'] = pd.to_datetime(df1['DATE'])
df1=df1.sort_values('DATE')
df1=df1.set_index('DATE')
df1
```

```python
df2 = pd.read_csv('KORFUNDS.csv')
df2['DATE'] = pd.to_datetime(df2['DATE'])
df2=df2.sort_values('DATE')
df2=df2.set_index('DATE')
```

```python
plt.figure(figsize=(19,6))
plt.rc('figure',titlesize=25)
plt.title(f'2019~2022년 한 미 금리')
plt.plot(df1, color='blue',label='미국')
plt.plot(df2, color='red',drawstyle='steps-post',label='한국')
plt.legend()
plt.text(df1.index,df1.iloc[0][13],df1.iloc[0][13],fontsize=15,color='black')
plt.show()
# plt.savefig('fund_chart_re.png')
```
